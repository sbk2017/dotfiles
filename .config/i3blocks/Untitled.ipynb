```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
lnk = 'http://api.huobi.pro/market/history/kline?period=1min&size=150&symbol=btcusdt'
req = requests.get(lnk).json()['data']
df = pd.DataFrame(req)
df
```

```python
sf = df.sort_values(by=['id'])
sf.reset_index(drop=True)
```

```python
sf['close'].tail(1)
```

```python
df['pandas_SMA_4'] = df.iloc[:,2].rolling(window=7).mean()
list(df['pandas_SMA_4'])
```
